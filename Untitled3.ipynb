## 1.5.10 : To check the percentage growth of sales for every month
we will see how much sales has incresed over 1 year for every month.
For this we have subtracted next year sale by previous year sale for the same month and shown it in percentage.<br>
((sale of April2018 - Sale of April 2017)/Sale of April 2017)*100

```python
#rename column name
df7.rename(columns={'Unnamed: 0':'Month'},inplace=True)
```

```python
#assigning data frame
df71=df7

#dropping row at index(12)
df71=df71.drop([12])
```

```python
#assigning columns to data frame
df77=df71[['FY15','FY16','FY17','FY18a']]
df77
```

```python
#assigning a new column in which calculated share value will be added
for index, row in df71.iterrows():   
        df71['16-15']=(df71['FY16']-df71['FY15'])/df7['FY15']*100
        df71['17-16']=(df71['FY17']-df71['FY16'])/df7['FY16']*100
        df71['18-17']=(df71['FY18a']-df71['FY17'])/df7['FY17']*100
```

**To display the graph for percentage share growth for the year 2017-18**

```python
#setting chart size
sns.set(rc={'figure.figsize':(22,8)})


#plotting bar chart
ax = sns.barplot(x="Month", y="18-17", data=df71)

#giving title 
plt.title("Month wise sale", fontsize=30)


#displaying chart
plt.show()
```

From the above graphical representation, we can see that sales in April 2018 are decreased as compared to sales in April 2017 (by 5.51%).

## 3.2 : Simple exponential smoothing
<ul>
<li>Exponential smoothing is a rule of thumb technique for smoothing time series data using the exponential window function.</li>
<li>Exponential Smoothing assigns exponentially decreasing weights as the observation get older.</li>
<li>Exponential smoothing is a way to smooth out data for presentations or to make forecasts. If you have a time series with a clear pattern, you could use moving averages but if you don't have a clear pattern you can use exponential smoothing to forecast.</li>
<li>The exponential smoothing calculation is as follows:</li>
</ul><br>
<b>F = (F-1) + D [(A-1) - (F-1)]</b><br>

**where,**<br>
          **F = Value to be forecast in future,**<br>
        **F-1 = Previously forecasted value,**<br>
          **D = Smoothing constant that varies between 0 < D < 1,**<br>
        **A-1 = Previous Actual Value.**<br>

### 3.2.1 : Model building 

```python
%matplotlib inline
#assigning values to variable
x=df881.values

#grouping  train and test dataset
test = [X[i] for i in range(0, 46)]


#definig function for exponential smoothing
def exponential_smoothing(series, alpha):
    result = [test[0]] # first value is same as series
    for n in range(1, len(test)):
        result.append(alpha * test[n] + (1 - alpha) * result[n-1])
    return result

#assigning values to variable
ta=exponential_smoothing(df881['Sales'],0.2)
tat=pd.DataFrame(ta)

#slicing series for calculation of error metrics
tata=tat[34:]
testa=test[34:]

#giving title
plt.title("Prediction VS Actual graph", fontsize=30)

#plotting graph
plt.ticklabel_format(useOffset=False, style='plain', axis='both')
pyplot.plot(tat , color='red',label='Prediction')
pyplot.plot(test,label='Actual')
plt.legend(loc='best')
pyplot.show()
```

P (AutoRegressive), D (Integrated) and Q (Moving Average) are the three properties of ARIMA model.
<ul>
    <li>Autuo Regression: A model that uses the dependent relationship between an observation and some number of lagged observations.
    </li>
     <li>Integrated: The use of differencing of raw observations(subtracting an observation from an observation at the previous time step) in order to make the time series stationary.

</ul><br>

where,<br>
p = number of lag observations included in the model.<br>
d = number of times that the raw observations are differenced.<br>
q = size of the moving average window<br>

```python
#defining function for MAPE
def mean_absolute_percentage_error(test, predictions): 
    test, predictions = np.array(test), np.array(predictions)
    return np.mean(np.abs((test - predictions) / test)) * 100

#definig function for MdAPE
def median_absolute_percentage_error(test, predictions): 
    test, predictions = np.array(test), np.array(predictions)
    return np.median(np.abs((test - predictions) / test)) * 100

#calculating MAE value
mae = mean_absolute_error(srmo, srmf)
print('MAE: %f' % mae)

#calculating MSE value
mse = mean_squared_error(srmo, srmf)
print('MSE: %f' % mse)

#calculating RMSE value
rmse = sqrt(mse)
print('RMSE: %f' % rmse)

#calculating MAPE value
mape=mean_absolute_percentage_error(srmo, srmf)
print('MAPE:', mape,"%")

#calculating MdAPE value
mdape=median_absolute_percentage_error(srmo, srmf)
print('MDAPE:', mdape,"%")

#appending model name to name list
name.append("seasonal_ARIMA")

#appending rmse value to error list
error.append(rmse)
```

The above values tell us the error terms i.e. we can decide how accurately our model is working.<br>
